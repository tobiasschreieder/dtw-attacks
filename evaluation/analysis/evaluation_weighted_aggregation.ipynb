```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, RegularPolygon
from matplotlib.path import Path
from matplotlib.projections.polar import PolarAxes
from matplotlib.projections import register_projection
from matplotlib.spines import Spine
from matplotlib.transforms import Affine2D
from cycler import cycler
import itertools
import zipfile

import json
import os
import copy
```

```python
def radar_factory(num_vars, frame='circle'):
    """Create a radar chart with `num_vars` axes.

    This function creates a RadarAxes projection and registers it.

    Parameters
    ----------
    num_vars : int
        Number of variables for radar chart.
    frame : {'circle' | 'polygon'}
        Shape of frame surrounding axes.

    """
    # calculate evenly-spaced axis angles
    theta = np.linspace(0, 2*np.pi, num_vars, endpoint=False)

    class RadarTransform(PolarAxes.PolarTransform):
        def transform_path_non_affine(self, path):
            # Paths with non-unit interpolation steps correspond to gridlines,
            # in which case we force interpolation (to defeat PolarTransform's
            # autoconversion to circular arcs).
            if path._interpolation_steps > 1:
                path = path.interpolated(num_vars)
            return Path(self.transform(path.vertices), path.codes)

    class RadarAxes(PolarAxes):

        name = 'radar'

        PolarTransform = RadarTransform

        def __init__(self, *args, **kwargs):
            super().__init__(*args, **kwargs)
            # rotate plot such that the first axis is at the top
            self.set_theta_zero_location('N')

        def fill(self, *args, closed=True, **kwargs):
            """Override fill so that line is closed by default"""
            return super().fill(closed=closed, *args, **kwargs)

        def plot(self, *args, **kwargs):
            """Override plot so that line is closed by default"""
            lines = super().plot(*args, **kwargs)
            for line in lines:
                self._close_line(line)

        def _close_line(self, line):
            x, y = line.get_data()
            # FIXME: markers at x[0], y[0] get doubled-up
            if x[0] != x[-1]:
                x = np.concatenate((x, [x[0]]))
                y = np.concatenate((y, [y[0]]))
                line.set_data(x, y)

        def set_varlabels(self, labels):
            self.set_thetagrids(np.degrees(theta), labels)

        def _gen_axes_patch(self):
            # The Axes patch must be centered at (0.5, 0.5) and of radius 0.5
            # in axes coordinates.
            if frame == 'circle':
                return Circle((0.5, 0.5), 0.5)
            elif frame == 'polygon':
                return RegularPolygon((0.5, 0.5), num_vars,
                                      radius=.5, edgecolor="k")
            else:
                raise ValueError("unknown value for 'frame': %s" % frame)

        def draw(self, renderer):
            """ Draw. If frame is polygon, make gridlines polygon-shaped """
            if frame == 'polygon':
                gridlines = self.yaxis.get_gridlines()
                for gl in gridlines:
                    gl.get_path()._interpolation_steps = num_vars
            super().draw(renderer)


        def _gen_axes_spines(self):
            if frame == 'circle':
                return super()._gen_axes_spines()
            elif frame == 'polygon':
                # spine_type must be 'left'/'right'/'top'/'bottom'/'circle'.
                spine = Spine(axes=self,
                              spine_type='circle',
                              path=Path.unit_regular_polygon(num_vars))
                # unit_regular_polygon gives a polygon of radius 1 centered at
                # (0, 0) but we want a polygon of radius 0.5 centered at (0.5,
                # 0.5) in axes coordinates.
                spine.set_transform(Affine2D().scale(.5).translate(.5, .5)
                                    + self.transAxes)


                return {'polar': spine}
            else:
                raise ValueError("unknown value for 'frame': %s" % frame)

    register_projection(RadarAxes)
    return theta

def intersection(lst1, lst2):
    return list(set(map(tuple, lst1)).intersection(map(tuple, lst2)))
```

```python
create_total = True

datasets = ["WESAD", "CGAN", "DGAN"]
dtw_attacks = ["single", "multi", "slicing", "multi-slicing"]

overall_data = {}
overall_matches = []
for dataset, dtw_attack in itertools.product(datasets, dtw_attacks):
    print(dataset, dtw_attack)

    with open(dataset + '_SW-DTW_evaluation_weightings_' + dtw_attack + '.json', 'r') as read_file:
        data = json.load(read_file)

    data_stress = data["stress"]["1"] + data["stress"]["3"] + data["stress"]["5"]

    stress_weights = list()
    for weights in data_stress:
        weight = list()
        for sensor in ['acc', 'bvp', 'eda', 'temp']:
            weight.append(weights[sensor])
        stress_weights.append(weight)

    data_non_stress = data["non-stress"]["1"] + data["non-stress"]["3"] + data["non-stress"]["5"]

    non_stress_weights = list()
    for weights in data_non_stress:
        weight = list()
        for sensor in ['acc', 'bvp', 'eda', 'temp']:
            weight.append(weights[sensor])
        non_stress_weights.append(weight)

    data = [['ACC', 'BVP', 'EDA', 'TEMP'], ('non-stress', non_stress_weights), ('stress', stress_weights)]

    # find match between all
    matches = intersection(non_stress_weights, stress_weights)
    overall_matches = intersection(overall_matches, matches) if overall_matches else matches
    print(overall_matches)

    # take all data per attack for total plot across datasets and stress and non-stress
    overall_data[dtw_attack] = [['ACC', 'BVP', 'EDA', 'TEMP'],
        (dtw_attack,
        overall_data[dtw_attack][1][1] + non_stress_weights + stress_weights
            if dtw_attack in overall_data else
        non_stress_weights + stress_weights
    )]

    if not create_total:
        N = len(data[0])
        theta = radar_factory(N, frame='polygon')

        plt.rc('font', size=14)
        yfontsize = 12
        plt.rcParams["axes.prop_cycle"] = plt.cycler("color", [plt.cm.tab20c.colors[0]])

        spoke_labels = data.pop(0)
        for title, case_data in data:
            fig, ax = plt.subplots(figsize=(3, 3), subplot_kw=dict(projection='radar'))
            fig.subplots_adjust(top=0.85, bottom=0.05)

            ax.set_rgrids([0.2, 0.4, 0.6, 0.8])
            ax.set_ylim(0, 1.0)
            ax.tick_params(axis='y', labelsize=8)
            plt.setp( ax.get_yticklabels(), visible=False)

            ax.text(0.3, 0.21, 0.2, color='black', fontsize=yfontsize, ha='center', va='center')
            ax.text(0.3, 0.38, 0.4, color='black', fontsize=yfontsize, ha='center', va='center')
            ax.text(0.3, 0.54, 0.6, color='black', fontsize=yfontsize, ha='center', va='center')
            ax.text(0.3, 0.7, 0.8, color='black', fontsize=yfontsize, ha='center', va='center')
            ax.text(0.3, 0.86, 1.0, color='black', fontsize=yfontsize, ha='center', va='center')
            #ax.set_title(title,  position=(0.5, 1.1), ha='center')

            for d in case_data:
                #line = ax.plot(theta, d)
                ax.fill(theta, d, alpha=0.25, label='_nolegend_')

            ax.set_varlabels(spoke_labels)

            plt.savefig("./weights_"+dataset+"_"+title+"_"+dtw_attack+".pdf", format="pdf", transparent=True, bbox_inches='tight')
            plt.show()

print("optimal weights for all attacks, datasets, and classes: ['ACC', 'BVP', 'EDA', 'TEMP'] - "+str(overall_matches))

if create_total:
    files = []
    for dtw_attack in dtw_attacks:
        print(dtw_attack)

        N = len(data[0])
        theta = radar_factory(N, frame='polygon')

        plt.rc('font', size=14)
        yfontsize = 12
        plt.rcParams["axes.prop_cycle"] = plt.cycler("color", [plt.cm.tab20c.colors[12]])

        spoke_labels = overall_data[dtw_attack].pop(0)
        for title, case_data in overall_data[dtw_attack]:
            fig, ax = plt.subplots(figsize=(3, 3), subplot_kw=dict(projection='radar'))
            fig.subplots_adjust(top=0.85, bottom=0.05)

            ax.set_rgrids([0.2, 0.4, 0.6, 0.8])
            ax.set_ylim(0, 1.0)
            ax.tick_params(axis='y', labelsize=8)
            plt.setp( ax.get_yticklabels(), visible=False)
            yfontsize = 12
            ax.text(0.3, 0.21, 0.2, color='black', fontsize=yfontsize, ha='center', va='center')
            ax.text(0.3, 0.38, 0.4, color='black', fontsize=yfontsize, ha='center', va='center')
            ax.text(0.3, 0.54, 0.6, color='black', fontsize=yfontsize, ha='center', va='center')
            ax.text(0.3, 0.7, 0.8, color='black', fontsize=yfontsize, ha='center', va='center')
            ax.text(0.3, 0.86, 1.0, color='black', fontsize=yfontsize, ha='center', va='center')
            #ax.set_title(title,  position=(0.5, 1.1), ha='center')

            for d in case_data:
                #line = ax.plot(theta, d)
                ax.fill(theta, d, alpha=0.05, label='_nolegend_')

            ax.set_varlabels(spoke_labels)

            savepath = "./weights_"+title+".pdf"
            files.append(savepath)
            plt.savefig(savepath, format="pdf", transparent=True, bbox_inches='tight')
            plt.show()

    with zipfile.ZipFile('weights_figures.zip', 'w', zipfile.ZIP_DEFLATED) as zipf:
        for entry in files:
            # Add each file to the ZIP archive
            zipf.write(entry, os.path.join('weights_figures', os.path.basename(entry)))
```
